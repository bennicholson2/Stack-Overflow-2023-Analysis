#### We are going to begin by looking at the yearly earnings of those working

##### I want to create a new dataframe where salary in usd is in a range of $ so that you can find a more accurate mode value

```python
min_salary_real = df[df['Salary (USD)'] > 100]

median_salary = df['Salary (USD)'].median()
mean_salary = df['Salary (USD)'].mean()
max_salary = df['Salary (USD)'].max(skipna = True)
min_salary_real = min_salary_real['Salary (USD)'].min(skipna = True)



basic_salary_info = pd.DataFrame({
    'Statistic': ['Median', 'Mean', 'Maximum', 'Minimum'],
    'Value': [median_salary, mean_salary, max_salary, min_salary_real]
    })
basic_salary_info = basic_salary_info.set_index('Statistic')
basic_salary_info['Value'] = basic_salary_info['Value'].round(decimals=0)
basic_salary_info
```

#### The dataframe series 'LanguageHaveWorkedWith' can have multiple values separated by semi-colons

#### The next goal is to create a counter for the frequency of any programming language in any response

```python
#Import the counter method
from collections import Counter
```

```python
language_counter = Counter()

for index, row in df.iterrows():
    if isinstance(row['LanguageHaveWorkedWith'], str):
        language_counter.update(row['LanguageHaveWorkedWith'].split(';'))
language_counter.most_common(10)
```

#### I want to know what individuals who are making over a million dollars a year have in common

```python
#First I need to make a filter which allows me to look at individuals who have a salary (USD) of over a million dollars
millionaire_df = df[df['Salary (USD)'] > 1000000.00]

#I am going to reuse the program from before to find the most commonly stated programming languages in their response
millionaire_language_counter = Counter()
for index, row in millionaire_df.iterrows():
    if isinstance(row['LanguageHaveWorkedWith'], str):
        millionaire_language_counter.update(row['LanguageHaveWorkedWith'].split(';'))

millionaire_language_counter.most_common(10)
```

```python
import pandas as pd

most_common_languages = millionaire_language_counter.most_common(5)

df_most_common = pd.DataFrame(most_common_languages, columns=['Language', 'Count'])
print(df_most_common)

plt.pie(df_most_common['Count'])
```
